# Univariate Analysis

```python
def uni_plot(feature):
    plt.figure(figsize=(4,3))
    plt.boxplot(df[feature], vert=False)  
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.show()
```

Scroll for checking univariate relations in below cell

```python
for col in df.columns:
    uni_plot(col)
```

```python
df['nonwhite'] = df['nonwhite'].map({0:'White', 1:'NonWhite'})
df['nonwhite'].value_counts()
```

```python
df['female'] = df['female'].map({0:'Male', 1:'Female'})
df['female'].value_counts()
```

```python
import random
color_palettes = [
    "deep",
    "muted",
    "bright",
    "pastel",
    "dark",
    "virdis",
    "Set1",
    "Set2",
    "Set3",
    "husl",
    "Blues",
    "Greens",
    "Oranges",
    "Reds",
    "Purples"
]
def bi_plot(feature1, feature2):
    plt.figure(figsize=(4, 3))
    choice = random.choice(color_palettes)
    sns.scatterplot(x=feature1, y=feature2, data=df, palette="choice")
    plt.title(f'Plot of {feature1} vs {feature2}')
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.plot()
```

# Bi-variate Analysis

Use the scroll bar in below cell to check bi-variate relations

```python
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        col_i = df.columns[i]
        col_j = df.columns[j]
        if (col_i == 'female' and col_j == 'nonwhite') or (col_i == 'nonwhite' and col_j == 'female'):
            continue
        bi_plot(col_i, col_j);
```
